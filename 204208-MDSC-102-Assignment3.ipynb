```python
# Plotting Special distribution functions.

import numpy as np
import math
from math import comb
import matplotlib.pyplot as plt
```

Binomial Distribution


$$B(n,p) = \sum_{j=0}^{x} \binom{n}{j} p^j (1 - p)^{n - j}$$

```python
# Plotting Binomial distribution.

def B(n,p):                    # n is the range of the r.v i.e:0,1,2,3,....,n
    x = np.arange(0,n+1,1)
    f = lambda x,n,p : comb(n,x)*(p**x)*((1-p)**(n-x)) #Lambda function for binominal distribution
    y = [f(i,n,p) for i in x]
    plt.bar(x,y, color='g')
    plt.plot(x,y, color='r')

    plt.title("Binomial Distribution")
    plt.xlabel("x")
    plt.ylabel("Probability")

B(20,0.5)         # Calling binomial distribution graph ploting function B(n,p): p = probaility, n= range of the r.v X

    
```

Poisson Distribution

$$Poi(λ)=e^{-λ}\frac{λ^x }{x!}$$



```python
# Plotting Poisson distribution.

def Poi(λ,n):                # n is the range of the r.v i.e:0,1,2,3,....,n
    x = np.arange(0,n+1,1)
    e = math.exp(-1*λ)
    f = lambda λ,x : e * ((λ**x)/math.factorial(x))      #Lambda function for poisson distribution
    y = [f(λ,i) for i in x]
    plt.bar(x,y)
    plt.plot(x,y, color='r')
    plt.title("Poisson Distribution")
    plt.xlabel("x")
    plt.ylabel("Probability")

Poi(1,15)   # Calling the poisson distribution graph ploting function Poi(λ,n): λ >= 0, n= range of the r.v X
```

Exponential Distribution

$$Exp(λ)=λe^{-λx}$$

```python
# Plotting Exponential distribution.

def Exp(λ):
    x = np.linspace(0,15,100)
    f = lambda λ,x : λ * math.exp(-1*λ*x)     # Lambda function for Exponential distribution.
    y = [f(λ,i) for i in x]
    plt.stem(x,y)
    plt.plot(x,y, color='r')

    plt.title("Exponential Distribution")
    plt.xlabel("x")
    plt.ylabel("Probability")

Exp(0.5) # Calling the Exponential distribution graph ploting function Exp(λ): λ >= 0
```

Normal Distribution

$$N(𝞵,σ^2)=\frac{1}{\sqrt{2\pi}σ}e^{-(x-𝞵)^2/2σ^2}$$

```python
# Plotting Normal distribution.
def N(μ,var,n):
    x = np.arange(-n,n,1)
    temp = 1/math.sqrt(2*3.14*var)
    f = lambda x,var,μ : temp * math.exp((-1/2)*(((x-μ)**2)/var))   # Lambda function for normal distribution.
    y = [f(i,var,μ) for i in x]
    plt.plot(x,y)
    plt.title("Normal Distribution")
    plt.xlabel("x")
    plt.ylabel("Probability")

N(1,100,30)     # Calling the normal distribution graph ploting function N(μ,var,n): where; n = the range of x(-n<=x<=n)
N(2,50,30)
```

Gamma Distribution

$$F(α,λ)=\frac{λ^α}{Γ(α)}x^{α-1}e^{-λx}$$

$$Γ(α)=(α-1)!$$

```python
# Plotting Gamma distribution.

def Gam(α,λ,n):
    x = np.linspace(0,n,1000)
    f = lambda x,λ,α : ((λ**α)*(x**(α-1))*math.exp(-1*λ*x))/math.factorial(α-1)   # Lambda function for gamma distribution.
    y = [f(i,λ,α) for i in x]
    plt.plot(x,y)
    plt.title("Gamma Distribution")
    plt.xlabel("x")
    plt.ylabel("Probability")

Gam(2,2,20)    # Calling the Gamma distribution graph ploting function Gam(α,λ,n): where; n = the range of x(0<=x<=n)
Gam(3,2,20)
Gam(5,0.5,20)
```

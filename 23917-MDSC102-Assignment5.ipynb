# Sub_graphs() Function :

     Sub_graphs(DataFrame,Groupby,x_axis,length,rows,cols) is a standard function for ploting scatter sub plots in which the points will be grouped by a particular feature and the x-axis will be constant for every graph. And, in each graph the y-axis will be a different feature.
     
     It will take 6 parameters :
     1) df = Is the pandas dataframe in the data is stores
     2) Groupby = Is the index of the feature(column) for which you want to group the scatter points
     3) x_axis = Is the index of the x-axis feature for the sub_graphs
     4) length = No.of features(columns) of the datasets
     5) rows = no.of rows in the graph
     6) cols = no.of columns in th graph

```python
def Sub_graphs(df,Groupby,x_axis,length,rows,cols):
    fig, axes = plt.subplots(rows, cols, figsize=(18, 21))
    fig.suptitle('Scatter Sub Plots Grouped by Class Features')

    temp,i,j = 0,0,0      # temp variable is for iterating over column index and i,j are for iterating for subgraph index
    while (temp < length):
        if ((temp != Groupby) and (temp != x_axis)):
            sns.scatterplot(ax = axes[i,j], x = df[df.columns[x_axis]] ,y= df[df.columns[temp]], hue = df[df.columns[Groupby]], palette='deep')
            j+=1
            if(j > 2):
                j = 0
                i += 1
        temp += 1
```

```python
Sub_graphs(df,0,1,14,4,3)  #Sub_graphs(DataFrame,Groupby,x_axis,length,rows,cols)
```

```python
def BoxSub_graphs(df,Groupby,x_axis,length,rows,cols):
    fig, axes = plt.subplots(rows, cols, figsize=(18, 21))
    fig.suptitle('Box Sub Plots Grouped by Class Features')

    temp,i,j = 0,0,0      # temp variable is for iterating over column index and i,j are for iterating for subgraph index
    while (temp < length):
        if ((temp != Groupby) and (temp != x_axis)):
            sns.boxplot(ax = axes[i,j], x = df[df.columns[x_axis]] ,y= df[df.columns[temp]], hue = df[df.columns[Groupby]], palette='deep')
            j+=1
            
            if(j > 2):
                j = 0
                i += 1
        temp += 1
```

# Plotting Histogram graph for every feature and each graph consists of only one Class.

```python
Class1 = df[df['Class'] == 1]     # Filtering Class1 data
Class2 = df[df['Class'] == 2]     # Filtering Class2 data
Class3 = df[df['Class'] == 3]     # Filtering Class3 data

def HistSub_graphs(x_axis):  # The input of the function is the feature(Column) name
    fig, axes = plt.subplots(1, 3, figsize=(15,5))
    fig.suptitle(f'Hist Sub Plots of {x_axis}')

    sns.histplot(ax = axes[0], data=Class1 ,x= x_axis , hue = 'Class',kde=True)
    sns.histplot(ax = axes[1], data=Class2 ,x= x_axis , hue = 'Class',kde=True)
    sns.histplot(ax = axes[2], data=Class3 ,x= x_axis , hue = 'Class',kde=True)

```

```python
for cols in df:     #iterating over feature(Column) name
    if(cols != 'Class'):
        HistSub_graphs(cols) # Calling the Function

# In the below graph every row represent the histogram graph of one of the feature's with respect to Class1,Class2 and Class3.
```
